```python
def read_schoolNames2016(paths):
    school_names= []
    for filename in paths:
        pdfReader = PdfFileReader(open(filename, 'rb'))
        pageObj = pdfReader.getPage(0)
        Text= pageObj.extractText()
        school_text= Text.splitlines()[0]+Text.splitlines()[1]
        
        school_names.append(school_text)
    
    return school_names
```

```python
def conditional_read2016(paths): #definition that detects pdfs with exactly 3 pages
    pdfs_with3 =[]
    for filename in paths:
        pdfReader2 = PdfFileReader(open(filename, 'rb'))
        NumPages= pdfReader2.getNumPages() #
        if NumPages == 3:
            pdfs_with3.append(filename)
        
    return pdfs_with3
```

```python
pdfs_with3pages= conditional_read2016(paths)
pdfs_with3pages
```

```python
#another definition that detects whether the last page actually has any content to extract
def conditional_3rdPage(pdfs_with3pages):
    PageContent3rdNone =[]
    for filename2 in pdfs_with3pages:
        
        PageContent3rd = tabula.read_pdf(
        filename2,lattice= True,guess= False, multiple_tables=True, 
        pages = '3') #Lattice= True is important and is required for detection
        
        if len(PageContent3rd) == 1: #An empty content is essentially a single blank entry in the list
            print("This file is empty" + filename2)
            PageContent3rdNone.append(filename2)
        
             
    return PageContent3rdNone
```

```python
#This list contains links of pdfs with 3 pages but no content on the last.
PageContent3rdNone = conditional_3rdPage(pdfs_with3pages)
PageContent3rdNone
len(PageContent3rdNone)
```
