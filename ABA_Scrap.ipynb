```python
class SelectFromDropDownMenu(unittest.TestCase):
    def Retrieve(self):
        global driver
        os.environ["PATH"] += os.pathsep + r'C:/Users/Sou/Documents/rocio/';
        driver= webdriver.Chrome()
        driver.get('http://abarequireddisclosures.org/Disclosure509.aspx')
    def SelectOptionYear(self):
        dropDownID = "ctl00_ContentPlaceHolder1_ddlYear"
        dropDownElement= WebDriverWait(driver,10).\
                         until(lambda driver: driver.find_element_by_id(dropDownID))
        Select(dropDownElement).select_by_visible_text("2011")
    def SelectSchool(self):
        dropDownSchool= "ctl00_ContentPlaceHolder1_ddlUniversity"
        dropDownElement= WebDriverWait(driver,10).\
                         until(lambda driver: driver.find_element_by_id(dropDownSchool))
        Select(dropDownElement).select_by_index(1)
    def Submit(self):
        SubmitButton = "ctl00_ContentPlaceHolder1_btnSubmit"
        ButtonClick = driver.find_element_by_id(SubmitButton)
        ButtonClick.click()

SelectFromDropDownMenu.Retrieve('self')
SelectFromDropDownMenu.SelectOptionYear('self')
SelectFromDropDownMenu.SelectSchool('self')
SelectFromDropDownMenu.Submit('self') 

#Works! Make sure to set the download folder as desired and allow automatic download.
#Also consider automating the ways the pdfs are named for better organization
```
